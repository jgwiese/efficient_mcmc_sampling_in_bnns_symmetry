```python
import matplotlib.pyplot as plt

def plot_spreads(k_ratios, values, spread_full, title):
    figure = plt.figure()
    ax = figure.add_subplot(1, 1, 1)
    ax.plot(k_ratios, values, label="reduced")
    ax.scatter(k_ratios, values)
    ax.axhline(y=spread_full, color='r', linestyle='-', label="full / before")
    ax.set_xlabel("k relative")
    ax.set_ylabel("spread / eigenvalues mean")
    ax.legend()
    ax.set_title(title)
    return figure
```

```python
# mlp 1, 4, 1, izmailov, rbf
spread_full = 5.222297
k_ratios = [0.001, 0.01, 0.125, 0.25, 0.5, 0.75, 1.0]
spread_reduced = [3.5101812, 0.22475415, 0.15480001, 0.15480001, 0.15480001, 0.15480001, 0.15480001]
plot_spreads(k_ratios, spread_reduced, spread_full, title="mlp 1, 4, 1; izmailov; rbf")
```

```python
# mlp 1, 4, 4, 1, izmailov, rbf
spread_full = 2.2368486
k_ratios = [0.001, 0.01, 0.125, 0.25, 0.5, 0.75, 1.0][:3]
spread_reduced = [1.556567, 1.3022218, 1.2371384]
plot_spreads(k_ratios, spread_reduced, spread_full, title="mlp 1, 4, 4, 1; izmailov; rbf")
```
