```python
%load_ext autoreload
%autoreload 2
%matplotlib agg

import os
import jax
import jax.numpy as jnp
import visualization
from utils import results, experiments, settings
import global_settings
from data import datasets
import matplotlib.pyplot as plt
```

```python
sinusoidal = datasets.Sinusoidal()
izmailov = datasets.Izmailov()
regression2d = datasets.Regression2d()
```

```python
textsize = 18
figure = plt.figure(figsize=(12, 4), tight_layout=True)
scale = 3

# (a)
ax1 = figure.add_subplot(1, 3, 1)
ax1.set_aspect("equal")
ax1.grid(visible=True)
ax1.set_xlim([-scale, scale])
ax1.set_ylim([-scale, scale])
ax1.set_xlabel(r"$x$")
ax1.set_ylabel(r"$y$")
ax1.xaxis.label.set_size(textsize)
ax1.yaxis.label.set_size(textsize)
ax1.tick_params(axis='both', which='major', labelsize=textsize - 2)
ax1.scatter(sinusoidal.data[:, 0], sinusoidal.data[:, 1], s=16, color="black", alpha=0.4, rasterized=True)

# (a)
ax2 = figure.add_subplot(1, 3, 2)
ax2.set_aspect("equal")
ax2.grid(visible=True)
ax2.set_xlim([-scale, scale])
ax2.set_ylim([-scale, scale])
ax2.set_xlabel(r"$x$")
ax2.set_ylabel(r"$y$")
ax2.xaxis.label.set_size(textsize)
ax2.yaxis.label.set_size(textsize)
ax2.tick_params(axis='both', which='major', labelsize=textsize - 2)
ax2.scatter(izmailov.data[:, 0], izmailov.data[:, 1], s=16, color="black", alpha=0.4, rasterized=True)

# (a)
ax3 = figure.add_subplot(1, 3, 3, projection="3d")
ax3.grid(visible=True)
ax3.set_xlim([-scale, scale])
ax3.set_ylim([-scale, scale])
ax3.set_xlabel(r"$x$")
ax3.set_ylabel(r"$y$")
ax3.xaxis.label.set_size(textsize)
ax3.yaxis.label.set_size(textsize)
ax3.tick_params(axis='both', which='major', labelsize=textsize - 2)
ax3.scatter(regression2d.data[:, 0], regression2d.data[:, 1], regression2d.data[:, 2], s=16, color="black", alpha=0.4, rasterized=True)

figure
```
