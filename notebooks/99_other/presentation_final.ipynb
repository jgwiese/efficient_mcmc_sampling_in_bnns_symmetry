```python
# tanh plane visualization
import matplotlib
import matplotlib.pyplot as plt
import ipyvolume as ipv
cmap = matplotlib.cm.get_cmap("gist_rainbow") # "jet"
import numpy as np
```

```python
tanh_sequential_helper = equioutput.SequentialHelper(
    transformation=experiment._model_transformation,
    dataset=experiment._dataset
)
tanh_structured_sequential_samples_parameters = tanh_sequential_helper.structured_sequential_samples_parameters(
    samples_parameters=result.samples["parameters"]
)
tanh_symmetry_helper = equioutput.SymmetryHelper(tanh_structured_sequential_samples_parameters)
```

```python
tanh_subspace_0 = tanh_symmetry_helper.hidden_layer_subspace(0)
```

```python
tanh_subspace_0_reduced = np.array(tanh_subspace_0)
for h in range(tanh_subspace_0.shape[1]):
    sele = tanh_subspace_0[:, h] @ svm._history[31][0] < 0
    tanh_subspace_0_reduced[sele, h] = -tanh_subspace_0_reduced[sele, h]
```

```python
def get_plane(normal):
    u, s, vh = jnp.linalg.svd(jnp.expand_dims(normal, 0))
    t = jnp.linspace(-6.0, 6.0, 8)
    xx, yy = jnp.meshgrid(t, t)
    plane = jnp.stack([xx, yy], -1) @ vh[-2:]
    return plane
```

```python
plane = get_plane(svm._history[31][0])
```

```python
fig = ipv.figure(width=512, height=512)
for h in range(tanh_subspace_0.shape[1]):
    color = cmap(1.0 * h / tanh_subspace_0.shape[1])
    scatter = ipv.scatter(
        tanh_subspace_0[:, h, 0],
        tanh_subspace_0[:, h, 1],
        tanh_subspace_0[:, h, 2],
        marker="sphere", color=color, size=0
    )
    scatter = ipv.scatter(
        tanh_subspace_0_reduced[:, h, 0],
        tanh_subspace_0_reduced[:, h, 1],
        tanh_subspace_0_reduced[:, h, 2],
        marker="sphere", color=color, size=2
    )
#scatter = ipv.scatter(x_reduced[:, 0], x_reduced[:, 1], x_reduced[:, 2], marker="sphere", color="blue", size=2)
surface = ipv.plot_wireframe(plane[..., 0], plane[..., 1], plane[..., 2], color="#000000")
ipv.pylab.style.axes_off()
#ipv.pylab.style.box_off()
ipv.view(azimuth=45, elevation=45, distance=2.2)
ipv.show()
```
